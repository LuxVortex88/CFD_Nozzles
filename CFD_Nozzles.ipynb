```python


import numpy as np
import matplotlib.pyplot as plt

# Define nozzle geometry as a 1D profile (simplified)
x = np.linspace(0, 1, 100)  # nozzle length normalized
area = 1 - 0.5 * x  # converging nozzle area profile

# Assume pressure and velocity relationship (simplified)
pressure = 101325 * (1 - 0.8 * x)  # pressure drops along nozzle
velocity = np.sqrt(2 * (101325 - pressure) / 1.225)  # velocity using Bernoulli (approx)

plt.plot(x, velocity)
plt.title("Velocity Profile in Rocket Nozzle (Simplified)")
plt.xlabel("Nozzle Length (normalized)")
plt.ylabel("Velocity (m/s)")
plt.show()
```
